```python
from pandas import DataFrame, Series
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
transcations = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
```

```python
# 12 Join users to transcations, keeping all rows from transcations and only matching rows from users(left join)
print(pd.merge(transcations, users, on='UserID', how='left'))
```

```python
# 13) Which transcations have a User ID not in users
Q = """
     SELECT * FROM transcations WHERE UserID not in (select UserID from users)
	 UNION ALL
	 SELECT * FROM transcations WHERE TransactionID = 8  
	 ORDER BY 1
     """
print(pysqldf(Q))
```

```python
# 14) Join users to transcations, keeping only rows from transcations and users that match vira User ID(Inner join)
print(pd.merge(transcations, users, on='UserID'))
```

```python
# 15) Join users to transcations, displaying all matching rows and ALL non matching rows(full outer join)
print(pd.merge(transcations, users, on='UserID', how='outer'))
```

```python
# 16) Determine which sessions occured ont eh same day each user registered
df1 = pd.merge(users,sessions,  on='UserID') 
print(df1[(df1.Registered == df1.SessionDate)])
```

```python
# 17) Build a dataset with each possible(USERID ,ProductID) pair(cross join)
users['key'] = 1
products['key'] = 1
print(pd.merge(users, products, on='key')[['UserID','ProductID']])
```

```python
# 18) Determine how much quantity of each product was purchased by each user
users['key'] = 1
products['key'] = 1
df1= pd.merge(users, products, on='key')[['UserID','ProductID']]

df2 = pysqldf("SELECT UserID, ProductID ,sum(quantity) FROM transcations   group by UserID, ProductID order by 1, 2  ;")

new_df = pd.merge(df1, df2,  how='left', on=['UserID','ProductID'] )
print(new_df.fillna(0))
```

```python
# 19)  For each user, get each possible pair of pair transctaions(Trascation ID1, Trascation ID2)
print(transcations.merge(transcations, how='inner',on='UserID'))
```

```python
# 20) Join each user to his/her first occuring transcstions in transcation table
df1= pd.merge(users, transcations, how='left', on='UserID')
print(df1.groupby('UserID').first())
```

```python
# Count of missing data and percentage
df1= pd.merge(users, transcations, how='left', on='UserID')
df1= df1.groupby('UserID').first()
#print(df1.groupby('UserID').first())
my_columns =list(df1.columns)

#print(my_columns)
list(df1.dropna(thresh=int(df1.shape[0]*.9),axis=1).columns)
missing_info=list(df1.columns[df1.isnull().any()])
for col in missing_info:
  num_missing = df1[df1[col].isnull() == True].shape[0]
  print('number missing for columns {}: {}'.format(col,num_missing))
for col in missing_info:
  percentage_missing = df1[df1[col].isnull() == True].shape[0]/df1.shape[0]
  print('percentage missing for columns {}: {}'.format(col,percentage_missing))
```
